```python
from scipy import stats
import matplotlib.pyplot as plt
import numpy.random as npr
```

### Bernoulli vs binomial

```python
npr.seed(42)
n = 50
bern = npr.binomial(1, 0.3, size=n)

plt.hist(bern)
plt.show()
```

```python
binom = npr.binomial(10, 0.5, size=1000)

plt.hist(binom)
plt.show()
```

### Beta distribution

```python
from scipy import stats
import numpy as np

fig, axes = plt.subplots(2,2, figsize=(10,6))
axes = axes.ravel()

for ax, (a,b) in zip(axes, [(0.1,0.1), (1,1), (2,3), (8,4)]):
    dom = np.linspace(0.01,0.99, 200)
    dist = stats.beta.pdf(dom, a, b)

    ax.plot(dom, dist, label=f'a={a}, b={b}')
    ax.legend()

plt.show()
```

### Simple Bayesian - one step

```python
# prior

a = 2
b = 2

prior = stats.beta.pdf(dom, a, b)
dom = np.linspace(0,1,200)

# likelihood function
N = 1
m = 1
LF = stats.binom.pmf(k=m, n=N, p=dom)

# posterior
posterior = LF*prior*(N+1)
# posterior2 = stats.beta.pdf(dom, m+a-1, N-m+b-1)
posterior2 = stats.beta.pdf(dom, a+1, b)

fig, axes = plt.subplots(1,3, figsize=(15,3))
axes = axes.ravel()

axes[0].plot(dom, prior)
axes[0].set_title('Prior')

axes[1].plot(dom, LF)
axes[1].set_title('Likelihood function')

axes[2].plot(dom, posterior, label='not normalized')
axes[2].set_title('posterior')
axes[2].plot(dom, posterior2, label='normalized')
axes[2].legend()

plt.show()
```

### multiple steps

```python
# prior
a = 3
b = 2
prior = stats.beta.pdf(dom, a, b)

# likelihood function
N = 2
m = 1
LF = stats.binom.pmf(k=m, n=1, p=dom)

# posterior
posterior = prior*LF*1.66
posterior2 = stats.beta.pdf(dom, a+1, b)

fig, axes = plt.subplots(1,3, figsize=(15,3))
axes = axes.ravel()

axes[0].plot(dom, prior, label=f'a={a}, b={b}')
axes[0].set_title('Prior')
axes[0].legend()

axes[1].plot(dom, LF)
axes[1].set_title('Likelihood function')

axes[2].plot(dom, posterior, label='not normalized')
axes[2].set_title(f'posterior, a={a+1}, b={b}')
axes[2].plot(dom, posterior2, label='normalized')
axes[2].legend()

plt.show()
```
